```julia
set = []
l = 50
for x ∈ 1:l
    for y ∈ 1:l
        if x > y
            X,Y = Int(x^2 - y^2), Int(2x*y)
            append!(set,[[X,Y]])
        end
    end
end
```

```julia
for i ∈ set
    print(i)
end
```

```julia
function normalise(x)
    t = 0
    for i in x
        t += i^2
    end
    return x/sqrt(t), sqrt(t)
end
```

```julia
u = []
v = []
U = []
V = []

norm_u = []
norm_v = []
RST = []

set2 = Set([]) #unique
for i ∈ set
    cd = gcd(i)
    push!(set2, [Int(i[1]/cd), Int(i[2]/cd)])
    push!(set2, [Int(i[2]/cd), Int(i[1]/cd)])
end


set3 = Set([]) #non-unique
for j ∈ set2
    for w ∈ 1:5
        push!(set3,[Int(w*j[1]),Int(w*j[2])])

    end
end
        


for k ∈ set2
    if k[1]^2 + k[2]^2 < l^4
        append!(u,k[1])
        append!(v,k[2])
    end
end

for st ∈ set3
    if st[1]^2 + st[2]^2 < l^4
        append!(U,st[1])
        append!(V,st[2])
    end
end

for t ∈ set2
    norming,rst = normalise(t)
    append!(norm_u, norming[1])
    append!(norm_v,norming[2])
    append!(RST,rst)
end

# for ts ∈ 1:length(norm_u)
#     print([ Int(round(RST[ts]*norm_u[ts],digits = 0)) , Int(round(RST[ts]*norm_v[ts],digits = 0))  ])
# end

# println()
# println()

# print(set2)


```
